# 2. Import Dependencies

```python
import cv2
import matplotlib.pyplot as plt
```

# 3. Connecting to camera

### 3.1 captures single frames

```python
# connect to capture device, try different values
cap = cv2.VideoCapture(0)
```

```python
# "frame" stores captured frame, "ret" stores True/False for connection
ret, frame = cap.read()
```

```python
# show captured frame
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)) 
plt.show()
```

```python
# releases captured frame
cap.release()
```

```python
def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('webcam_photo.jpg', frame)
    cap.release()
```

```python
take_photo()
```

### 3.2 Real time capture

```python
# Connect to webcame
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FOURCC, cv2.VideoWriter_fourcc(*"MJPG"))

# Loop through every frame until webcam is closed
while cap.isOpened():
    ret, frame = cap.read()
    
    cv2.imshow('webcam', frame)
    
    # Detects for key press and then check if 'q' was pressed
    if cv2.waitKey(1) & 0xFF == ord('q)'):
        break
cap.release()
cv2.destroyAllWindows()
```
